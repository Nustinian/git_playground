```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
url = "http://pyclass.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/"
req = requests.get(url, headers = headers)
content = req.content
soup = BeautifulSoup(content, "html.parser")

pages = soup.find_all("span", {"id": "PageNumbers", "class": "PageNumbers"})[0]
pages = [url] + [url + a['href'] for a in pages.find_all("a", href = True) if "javascript" not in a['href']]

prop_dict_list = []
for page in pages:
    req = requests.get(page, headers = headers)
    content = req.content
    soup = BeautifulSoup(content, "html.parser")
    all = soup.find_all("div",{"class": "propertyRow"})
    for property in all:
        prop_dict = {}
        prop_dict["address"] = property.find_all("span", {"class": "propAddressCollapse"})[0].text + ", " + property.find_all("span", {"class": "propAddressCollapse"})[1].text
        prop_dict["price"] = property.find("h4", {"class": "propPrice"}).text.replace("\n", "").replace(" ", "")
        try:
            prop_dict["beds"] = property.find("span", {"class": "infoBed"}).find("b").text
        except:
            prop_dict["beds"] = None
        try:
            prop_dict["fullbaths"] = property.find("span", {"class": "infoValueFullBath"}).find("b").text
        except:
            prop_dict["fullbaths"] = None
        try:
            prop_dict["sqft"] = property.find("span", {"class": "infoSqFt"}).find("b").text
        except:
            prop_dict["sqft"] = None
        try:
            prop_dict["halfbaths"] = property.find("span", {"class": "infoValueHalfBath"}).find("b").text
        except:
            prop_dict["halfbaths"] = None
        
        cgroup = property.find_all("div", {"class": "columnGroup"})
        for c in cgroup:
            if c.find("span", {"class": "featureGroup"}):
                if "Lot Size:" in c.find("span", {"class": "featureGroup"}).text:
                    lotsize = [span.text for span in c.find_all("span", {"class": "featureName"})]
                    prop_dict["lotsize"] = "".join(lotsize)
            else:
                prop_dict["lotsize"] = None
        prop_dict_list.append(prop_dict)
```

```python
import pandas
data = pandas.DataFrame(prop_dict_list)
```
